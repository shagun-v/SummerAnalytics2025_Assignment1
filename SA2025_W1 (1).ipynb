### 1. What is name of car that has the highest horsepower?  

```python
# Reload the original dataset to get names back
cars = pd.read_csv('Cars.csv')
max_hp_row = cars.loc[cars['horsepower'].idxmax()]
print(f"The car with the highest horsepower is '{max_hp_row['name']}' with {max_hp_row['horsepower']} HP")
```

### 2. How many cars have mpg  35?

```python
# Count cars with mpg >= 35
high_mpg_count = len(cars[cars['mpg'] >= 35])

print(f"Number of cars with MPG  35: {high_mpg_count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Filter cars meeting both conditions
filtered_cars = cars[(cars['horsepower'] > 100) & (cars['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_cars['origin'].mode()[0]

# Count occurrences
origin_counts = filtered_cars['origin'].value_counts()

print(f"The most common origin is: {most_common_origin}")
print("\nBreakdown of origins:")
print(origin_counts)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Filter Japanese cars and calculate mean acceleration
japanese_accel = cars[cars['origin'] == 'japan']['acceleration'].mean()

# Round to 2 decimal places
rounded_accel = round(japanese_accel, 2)

print(f"Mean acceleration of Japanese cars: {rounded_accel}")
```

### 5. Which year had the highest average mpg?

```python
# Calculate average MPG by year
avg_mpg_by_year = cars.groupby('model_year')['mpg'].mean()

# Find the year with highest average MPG
best_year = avg_mpg_by_year.idxmax()
highest_avg_mpg = round(avg_mpg_by_year.max(), 2)

print(f"The year with highest average MPG is {best_year} with {highest_avg_mpg} MPG")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Calculate median MPG
median_mpg = cars['mpg'].median()

# Filter cars with above-median MPG
high_mpg_cars = cars[cars['mpg'] > median_mpg].copy()

# Calculate hp/weight ratio if not already present
if 'hp_to_weight' not in high_mpg_cars.columns:
    high_mpg_cars['hp_to_weight'] = high_mpg_cars['horsepower'] / high_mpg_cars['weight']

# Find car(s) with maximum hp/weight ratio
max_ratio = high_mpg_cars['hp_to_weight'].max()
best_cars = high_mpg_cars[high_mpg_cars['hp_to_weight'] == max_ratio]

# Display results
print(f"Median MPG: {median_mpg:.1f}")
print(f"Maximum hp/weight ratio among high-MPG cars: {max_ratio:.6f}")
print("\nCar(s) with best ratio:")
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_to_weight']].to_string(index=False))
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Calculate average MPG by year and origin
avg_mpg = cars.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

# Create plot
ax = sns.lineplot(data=avg_mpg, dashes=False, markers=True, markersize=8)

# Customize plot
plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=14, pad=20)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.xticks(avg_mpg.index)
plt.grid(True, alpha=0.3)

# Add legend
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style and figure size
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Create the scatterplot
scatter = sns.scatterplot(
    data=cars,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    palette={'japan': 'green', 'europe': 'blue', 'usa': 'red'},
    sizes=(40, 200),  # Range of point sizes
    alpha=0.8
)

# Customize plot appearance
plt.title('Vehicle Weight vs Horsepower by Origin (Size = MPG)', fontsize=16, pad=20)
plt.xlabel('Horsepower (HP)', fontsize=14)
plt.ylabel('Weight (lbs)', fontsize=14)

# Improve legend
handles, labels = scatter.get_legend_handles_labels()
plt.legend(
    handles[:4],  # Show only origin and size legends
    labels[:4],
    title='Origin / MPG',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

# Add grid and adjust layout
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Show plot
plt.show()
```
